```python
%matplotlib
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
# hyper bolic coordinates
nu = 1
mu = np.linspace(-2*np.pi, 2*np.pi, 100)

sigma = 100
tau = np.linspace(-1, 1, 100)

a = 1

#make x and y
# x = a * np.cosh(mu) * np.cos(nu)
# y = a * np.sinh(mu) * np.sin(nu)
x = a * sigma * tau
y = np.sqrt(a**2 * (sigma**2 - 1) * (1 - tau**2)) 

sample = np.random.randint(0, 100, 100)

```

```python
# plot
plt.figure()
plt.plot(x, y, '-bs')
plt.plot(x[sample], y[sample], 'rs')
plt.show()
```
